split the data into train and test 

```python
from sklearn.model_selection import train_test_split
x=heart.drop(columns="output")
y=heart["output"]
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.20, stratify=y,random_state=42)
```

```python
print("x_train",x_train.shape)
print("y_train",y_train.shape)
print("x_test",x_test.shape)
print("y_test",y_test.shape)
```

sucessfully split the data into train and test

applying some model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import  KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import  MLPClassifier
from sklearn.svm import SVC
```

```python
LR = LogisticRegression()
DT = DecisionTreeClassifier()
RF = RandomForestClassifier()
KN = KNeighborsClassifier()
MLP = MLPClassifier()
SVR=SVC()
```

```python
LR.fit(x_train,y_train)
x_train_pred=LR.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score,r2_score
training_data_accu=accuracy_score(x_train_pred,y_train)
print('traning data accuracy',training_data_accu)
```

```python
x_test_pred=LR.predict(x_test)
testing_data_accu=accuracy_score(x_test_pred,y_test)
print('testing data accuracy',testing_data_accu)
```
